```python
from conllu import *
```

```python
with open('yo_ytb-ud-test.conllu') as f:
    contents = f.readlines()
```

```python
contents = "".join(contents)
contents
```

```python
sentences = parse(contents)
sentences
```

```python
look_for = ["ni"]
filtered_ss = [s for s in sentences if s.filter(form="ni")]
filtered_ss
```

```python
trees = [s.to_tree() for s in filtered_ss]
trees[11].token['form']
```

```python
def find_ni(t):
    lst = []
    if not t.children:
        return
    for c in t.children:
        if c.token['form'] == 'ni':
            return t
    for c in t.children:
        f = find_ni(c)
        if f:
            return f

for i in range(len(filtered_ss)):
    t = trees[i]
    ni = find_ni(t)
    if len(ni.children) == 1 and 'text_en' in t.metadata:
        ni.print_tree()
        print(t.metadata['text'])
        print(t.metadata['text_en'])
        print()
    
```

```python
pronouns = ["Èmi", "èmi", "Àwa", "àwa", "Ìwọ", "ìwọ", "Ẹ̀yin", "ẹ̀yin", "Òun", "òun", "Àwọn", "àwọn"]
ni_ss = []
for s in sentences:
    lst = s.metadata["text"].split(" ")
    if "ni" in lst:
        ind = lst.index("ni")
        if (lst[ind - 1] in pronouns or lst[ind + 1] in pronouns) and 'text_en' in s.metadata:
            print(s.metadata["text"])
            print(s.metadata["text_en"], end="\n\n")
```
